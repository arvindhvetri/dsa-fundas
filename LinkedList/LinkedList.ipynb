```python
# 1️⃣ Singly Linked List – MUST-KNOW Operations
# These are non-negotiable for placements.
# 🔹 Basic Operations (they expect instant answers)
# You should be able to code without thinking:
# Insert
    # At beginning
    # At end
    # At a given position
# Delete
    # From beginning
    # From end
    # Delete a given value
    # Delete at a given position
# Traversal / Print
    # Iterative
    # Recursive (bonus)
# Search
    # Iterative
    # Recursive
# 🔹 Pointer Logic / Core Interview Favorites
# These come up very frequently:
# Reverse a linked list
    # Iterative (MOST IMPORTANT)
    # Recursive (good to know)
# Find middle of linked list
    # Slow & Fast pointer
# Detect loop / cycle
    # Floyd’s Cycle Detection
    # Also: find start of loop
# Length of linked list
    # Iterative
    # Recursive
# 🔹 Advanced but High-Value Questions
# These separate average vs strong candidates:
# Remove duplicates
    # Sorted list
    # Unsorted list
# Merge two sorted linked lists
# Find nth node from end
    # Two-pointer approach
# Check if linked list is palindrome
    # Reverse second half
    # OR using stack
# Intersection point of two linked lists
# Delete node without head pointer
    # Classic trick question

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None
        self.tail = None

    def ins_beg(self, data):
        newNode = Node(data)

        # Empty list
        if self.head is None:
            self.head = newNode
            self.tail = newNode
            return

        # Non-empty list
        newNode.next = self.head
        self.head = newNode

    def ins_end(self, data):
        newNode = Node(data)
        if self.tail is None:
            self.head = newNode
            self.tail = newNode
            return

        self.tail.next = newNode
        self.tail = newNode

    def ins_loc(self, data, loc):
        llc = self.ll_count()
    
        # Invalid location
        if loc < 1 or loc > llc + 1:
            return "Enter Correct Location"
    
        # Insert at beginning
        if loc == 1:
            self.ins_beg(data)
            return
    
        # Insert at end
        if loc == llc + 1:
            self.ins_end(data)
            return
    
        # Insert in middle
        newNode = Node(data)
        tmp = self.head
        counter = 1
    
        while counter < loc - 1:
            tmp = tmp.next
            counter += 1
    
        newNode.next = tmp.next
        tmp.next = newNode

    def rm_beg(self):
        if self.head is None:
            return "No element in List"
    
        if self.head == self.tail:   # only one node
            self.head = None
            self.tail = None
            return
    
        self.head = self.head.next

    def rm_end(self):
        if self.tail is None:
            return "No element in List"
    
        # Only one node
        if self.head == self.tail:
            self.head = None
            self.tail = None
            return
    
        # Traverse to second-last node
        tmp = self.head
        while tmp.next != self.tail:
            tmp = tmp.next
    
        tmp.next = None
        self.tail = tmp

    def rm_loc(self, loc):
        llc = self.ll_count()
    
        # Invalid location
        if loc < 1 or loc > llc:
            return "Enter Correct Location"
    
        # Remove first node
        if loc == 1:
            self.rm_beg()
            return
    
        # Remove last node
        if loc == llc:
            self.rm_end()
            return
    
        # Remove middle node
        tmp = self.head
        counter = 1
        while counter < loc - 1:
            tmp = tmp.next
            counter += 1
    
        tmp.next = tmp.next.next

    def reverse(self):
        prev = None
        curr = self.head
        self.tail = self.head
    
        while curr is not None:
            nxt = curr.next
            curr.next = prev
            prev = curr
            curr = nxt
    
        self.head = prev

        

    def ll_count(self):
        llc = 0
        tmp = self.head
        while tmp is not None:
            llc += 1
            tmp = tmp.next
        return llc

    def display(self):
        res = []
        tmp = self.head
        while tmp is not None:
            res.append(tmp.data)
            tmp = tmp.next
        return res


l1 = LinkedList()
l1.ins_beg(5)
l1.ins_beg(4)
l1.ins_end(6)
l1.ins_end(7)
l1.ins_beg(3)
l1.ins_loc(2.5,1)
l1.reverse()
val = l1.display()
c = l1.ll_count()
print(val)
print(c)
```

```python
# Doublly LinkedList
class Node:
    def __init__(self,data):
        self.data = data
        self.prev = None
        self.next = None

class DLL:
    def __init__(self):
        self.head = None
        self.tail = None

    def ins_beg(self,data):
        newNode = Node(data)
        if self.head is None:
            self.head = newNode
            self.tail = newNode
            return

        newNode.next = self.head
        self.head.prev = newNode
        self.head = newNode

    def ins_end(self,data):
        newNode = Node(data)
        if self.tail is None:
            self.head = newNode
            self.tail = newNode
            return

        newNode.prev = self.tail
        self.tail.next = newNode
        self.tail = newNode

    def ins_loc(self,data,loc):
        newNode = Node(data)
        dllc = self.dllc()
        if loc < 1 or loc > dllc + 1:
            print("Enter correct location")
            return
        if loc == 1:
            self.ins_beg(data)
            return
        if loc == dllc + 1:
            self.ins_end(data)
            return
        counter = 1
        curr = self.head
        while counter < loc - 1:
            counter += 1
            curr = curr.next
        newNode.prev = curr
        newNode.next = curr.next
        curr.next.prev = newNode
        curr.next = newNode

    def del_beg(self):
        if self.head is None:
            print("No element in list")
            return
        if self.head == self.tail:
            self.head = None
            self.tail = None
            return
        self.head = self.head.next
        self.head.prev = None

    def del_end(self):
        if self.tail is None:
            print("No element in list")
            return
        if self.head == self.tail:
            self.head = None
            self.tail = None
            return
        self.tail = self.tail.prev
        self.tail.next = None

    def del_loc(self,loc):
        dllc = self.dllc()
        if loc<1 or loc > dllc:
            print("Enter crt location")
            return
        if loc == 1:
            self.del_beg()
            return
        if loc == dllc:
            self.del_end()
            return
        counter = 1
        curr = self.head
        while counter < loc :
            counter += 1
            curr = curr.next
        curr.prev.next = curr.next
        curr.next.prev = curr.prev

    def ft(self):
        tmp = self.head
        while tmp is not None:
            print(tmp.data,end=" ")
            tmp = tmp.next

    def bt(self):
        tmp = self.tail
        while tmp is not None:
            print(tmp.data,end=" ")
            tmp = tmp.prev
            
            
    def dllc(self):
        counter = 0
        curr = self.head
        while curr is not None:
            counter += 1
            curr = curr.next
        return counter
        
    def display(self):
        curr = self.head
        while curr is not None:
            print(curr.data,end=" ")
            curr = curr.next

dl1 = DLL()
dl1.ins_beg(4)
dl1.ins_beg(5)
dl1.ins_beg(3)
dl1.ins_end(6)
dl1.ins_end(7)
dl1.ins_loc(2,5)
dl1.display()
print("\n")
dl1.del_beg()
dl1.display()
print("\n")
dl1.del_end()
dl1.display()
print("\n")
dl1.del_loc(3)
#print(dl1.dllc())
dl1.display()
print("\n")
dl1.ft()
print("\n")
dl1.bt()
        
```
