```python
# Linear Search :-
# Start from first element
# Compare each element one by one
# Stop when the element is found or reach the end
# works both on sorted and unsorted list

l1 = [45, 22, 67, 22, 89, 14, 22, 90, 3]

def ls(l,t):
    for i, v in enumerate(l):
        if t == v:
            return i
    return -1

res = ls(l1,22)
print(res)
```

```python
# Binary Search :-
# Divide and Conquer process
# Works only on sorted arrays
# Find the middle element and compare with target

l1 = [3, 8, 15, 23, 42, 56, 67, 81, 94]

def bs(l,t):
    left = 0
    right = len(l) - 1
    while left<= right:
        mid = (left + right) // 2
        if t == l[mid]:
            return mid
        if t > l[mid]:
            left = mid + 1
        if t < l[mid]:
            right = mid - 1
    return -1
    
res = bs(l1,81)
print(res)
```
